```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'str']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'str']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
#Distribution of Numerical Features
df[numeric_features].hist(figsize=(18, 12), bins=20)

plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap
import seaborn as sns
plt.figure(figsize=(14,10))

sns.heatmap(df[numeric_features].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title("Correlation Heatmap")
```

```python
#Calories Burned by Workout Type
plt.figure(figsize=(10,6))

sns.boxplot(x='Workout_Type',
            y='Calories_Burned',
            data=df)

plt.title("Calories Burned by Workout Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#BMI vs Calories Burned
plt.figure(figsize=(8,6))

sns.scatterplot(x='BMI',
                y='Calories_Burned',
                hue='Gender',
                data=df)

plt.title("BMI vs Calories Burned")
plt.show()
```
